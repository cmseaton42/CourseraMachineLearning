```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.optimize import minimize

from sklearn.preprocessing import PolynomialFeatures
```

```python
def loaddata(file, delimeter = ","):
    data = np.loadtxt(file, delimiter=delimeter)
    print('Dimensions: ',data.shape)
    print(data[0:5,:])
    return(data)
```

```python
# Hypthosesis - Logistic/Sigmoid Function
def sigmoid(z):
    return (1 / (1 + np.exp(-z)))
```

```python
# Cost Function
def costFunction(theta, X, y):
    m = len(X)
    theta = np.c_[theta]
    
    z = X.dot(theta) # X * theta
    h = sigmoid(z) # Hypothesis
    
    cost = (1.0 / m) * (-y.T.dot(np.log(h)) - (1 - y).T.dot(np.log(1 - h)))
    
    return cost.flatten()[0]
```

```python
def costFunctionReg(theta, X, y, reg):
    m = len(X)
    theta = np.c_[theta]
    
    z = X.dot(theta) # X * theta
    h = sigmoid(z) # Hypothesis
    
    cost = (1.0 / m) * (-y.T.dot(np.log(h)) - (1 - y).T.dot(np.log(1 - h)))
    cost += (reg / (2 * m)) * (theta.T.dot(theta) - theta[0] ** 2)
    
    return cost.flatten()[0]
```

```python
def gradient(theta, X, y):
    m = len(X)
    theta = np.c_[theta]
    
    z = X.dot(theta) # X * theta
    h = sigmoid(z) # Hypothesis
    
    grad = (1.0 / m) * (X.T.dot(h - y))
    
    return grad.flatten()
```

```python
def gradientReg(theta, X, y, reg):
    m = len(X)
    theta = np.c_[theta]
    
    zeroFirst = np.ones((len(theta), 1))
    zeroFirst[0, 0] = 0
    
    z = X.dot(theta) # X * theta
    h = sigmoid(z) # Hypothesis
    
    grad = (1.0 / m) * (X.T.dot(h - y))
    grad += (reg / m) * theta * zeroFirst
    
    return grad.flatten()
```

```python
def predict(theta, X, threshold=0.5):
    return sigmoid(X.dot(theta)) >= threshold
```

```python
data = loaddata("./data/ex2data1.txt")
```

```python
m = data.shape[0] # num of training samples
n = data.shape[1] # num of features

X = np.c_[np.ones(m), data[:, 0:(n - 1)]]
y = np.c_[data[:, n - 1]]

admit = y[:, 0] == 1
nadmit = np.invert(admit)

fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.set_title("Exam 1 vs. Exam 2")
ax.set_xlabel("Exam 1 Score")
ax.set_ylabel("Exam 2 Score")
ax.scatter(X[admit][:, 1], X[admit][:, 2], s=30, c="g", marker="+", linewidth=1, label="Admitted")
ax.scatter(X[nadmit][:, 1], X[nadmit][:, 2], s=30, c="r", marker="x", linewidth=0.8, label="Not Admitted")
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
```

```python
initial_theta = np.zeros(n)
initial_cost = costFunction(initial_theta, X, y)
initial_grad = gradient(initial_theta, X, y)

print("Expected Cost of Initial Theta (Zeros):  0.693")
print("Actual   Cost of Initial Theta (Zeros): {0: 5.3f}".format(initial_cost))
print("Expected Gradient of Initial Theta (Zeros): [-0.100 -12.009 -11.263]")
print("Actual   Gradient of Initial Theta (Zeros): [{0: 5.3f} {1: 5.3f} {2: 5.3f}]".format(initial_grad[0], initial_grad[1], initial_grad[2]))
```

```python
%%time
initial_theta = np.zeros(n)

options = {
    "maxiter": 400
}

result = minimize(costFunction, initial_theta, args=(X,y), method="TNC", jac=gradient, options=options)
model = result.x

print("Found Regression Model: {} \n".format(model))
print("                      Optimization Result\n", result, "\n") 
```

```python
# Predict if a student with Exam Score 1 of 45 and Exam Score 2 of 85
#   be admitted using regression model parameters
probability = sigmoid(np.array([1, 45, 85]).dot(model.T))

print("Expected Chance a Student will be Admitted: 77.50%")
print("Actual   Chance a Student will be Admitted: {0:5.2f}%".format(probability * 100))
```

```python
p = predict(model, X)

print("Expected Train Accuracy: 89.00%")
print("Actual   Train Accuracy: {0:5.2f}%".format(100 * np.mean(p == y.ravel())))
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.set_title("Exam 1 vs. Exam 2")
ax.set_xlabel("Exam 1 Score")
ax.set_ylabel("Exam 2 Score")

# Data Scatter Plot
ax.scatter(X[admit][:, 1], X[admit][:, 2], s=30, c="g", marker="+", linewidth=1, label="Admitted")
ax.scatter(X[nadmit][:, 1], X[nadmit][:, 2], s=30, c="r", marker="x", linewidth=0.8, label="Not Admitted")
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)

# Get Min and Max of Exam 1 and Exam 2
xmin, xmax = X[:,1].min(), X[:,1].max()
ymin, ymax = X[:,2].min(), X[:,2].max()

# Build point matrix to sample against
#   trained model parameters
x = np.linspace(xmin, xmax, 100)
y = np.linspace(ymin, ymax, 100)
xx, yy = np.meshgrid(x, y)

# Use point matrix to sample sigmoid w/ 
#    trained model parameters 
mesh_Z = np.c_[np.ones((len(xx.flatten()), 1)), xx.flatten(), yy.flatten()]
hypothesis = sigmoid(mesh_Z.dot(model)).reshape(xx.shape)

# Gererate Descision Boundary
plt.contour(xx, yy, hypothesis, [0.5], linewidths=0.5, colors="b")
```

```python
data = loaddata("./data/ex2data2.txt")
```

```python
m = data.shape[0] # num of training samples
n = data.shape[1] # num of features

X = np.c_[data[:, 0:(n - 1)]]
y = np.c_[data[:, n - 1]]

one = y[:, 0] == 1
zero = np.invert(one)

fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.set_title("Test 1 vs. Test 2")
ax.set_xlabel("Microchip Test 1")
ax.set_ylabel("Microchip Test 2")
ax.scatter(X[one][:, 0], X[one][:, 1], s=30, c="g", marker="+", linewidth=1, label="y = 1")
ax.scatter(X[zero][:, 0], X[zero][:, 1], s=30, c="r", marker="x", linewidth=0.8, label="y = 0")
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
```

```python
poly_6 = PolynomialFeatures(6)
XX = poly_6.fit_transform(X)
print("New Feature Set Matrix Shape: {}".format(XX.shape))
```

```python
initial_theta = np.zeros(XX.shape[1])
initial_cost = costFunctionReg(initial_theta, XX, y, 1)
initial_grad = gradientReg(initial_theta, XX, y, 1)

print("Expected Cost of Initial Theta (Zeros):  0.693")
print("Actual   Cost of Initial Theta (Zeros): {0: 5.3f}".format(initial_cost))
print("Expected Gradient of Initial Theta (Zeros) (First 4 Only): [0.0085 0.0188 0.0001 0.0503]")
print("Actual   Gradient of Initial Theta (Zeros) (First 4 Only): [{0:6.4f} {1:6.4f} {2:6.4f} {3:6.4f}]"
      .format(initial_grad[0], initial_grad[1], initial_grad[2], initial_grad[3]))
```

```python
%%time
initial_theta = np.zeros(XX.shape[1])

fig, axes = plt.subplots(1, 3, figsize=[16, 5])
lambdas = [0, 1, 100]

for i, value in enumerate(lambdas):
    ax = axes[i]
    
    opt = {
        "maxiter": 3000
    }
    
    # Find Optimal Model Parameters
    result = minimize(costFunctionReg, initial_theta, args=(XX, y, value), method="TNC", jac=gradientReg, options=opt)
    model = result.x
    
    # Find Accuracy
    accuracy = 100 * np.mean(predict(model, XX) == y.ravel())

    # Data Scatter Plot
    ax.scatter(X[one][:, 0], X[one][:, 1], s=30, c="g", marker="+", linewidth=1, label="y = 1")
    ax.scatter(X[zero][:, 0], X[zero][:, 1], s=30, c="r", marker="x", linewidth=0.8, label="y = 0")
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles, labels)

    # Get Min and Max of Test 1 and Test 2
    xmin, xmax = X[:,0].min(), X[:,0].max()
    ymin, ymax = X[:,1].min(), X[:,1].max()

    # Build point matrix to sample against
    #   trained model parameters
    x_arr = np.linspace(xmin, xmax, 100)
    y_arr = np.linspace(ymin, ymax, 100)
    xx, yy = np.meshgrid(x_arr, y_arr)

    # Use point matrix to sample sigmoid w/ 
    #    trained model parameters 
    mesh_Z = poly_6.fit_transform(np.c_[xx.flatten(), yy.flatten()])
    hypothesis = sigmoid(mesh_Z.dot(model)).reshape(xx.shape)

    # Gererate Descision Boundary
    ax.contour(xx, yy, hypothesis, [0.5], linewidths=0.5, colors="b")
    
    # Add Labels
    ax.set_title(r"Accuracy: {0:5.2f}% , $\lambda$ = {1}".format(accuracy, value))
    ax.set_xlabel("Microchip Test 1")
    ax.set_ylabel("Microchip Test 2")
```
