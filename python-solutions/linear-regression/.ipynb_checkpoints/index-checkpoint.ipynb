```python
# Add Main Libraries to Global Namespace
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import axes3d
```

```python
# Read in Data File 1
raw_data_1 = open("./data/ex1data1.txt")
data_1 = np.loadtxt(raw_data_1, delimiter=",").astype("float")
```

```python
# Parse Data_1 File

# Get Training Data Length
m = len(data_1)

# Get Training Features
X_Unbiased = np.c_[data_1[:, 0]]

# Get Training Results
Y = np.c_[data_1[:, 1]]

# Initialize Plot Object
fig = plt.figure(figsize=(12,8))

# Generate Plot
ax = fig.add_axes([0, 0, .42, .42])
ax.set_title("Population vs. Profit")
ax.set_xlabel("Population of City in 10,000s")
ax.set_ylabel("Profit in $10,000s")
ax.scatter(X_Unbiased, Y, s=30, c="b", marker="x", linewidth=0.9)
```

```python
# Prep to Run Gradient Descent

# Inject Bias Row into Feature Set and Build Column Vectors
X = np.c_[np.ones(m), X_Unbiased]

# Initialize Parameters
theta = np.zeros((2, 1))

# Gradient Descent Options
iterations = 1500
alpha = 0.01 # learning rate
```

```python
# Define Cost Function
def computeCost(X, y, m, theta):
    return (0.5 / m) * np.sum((np.matmul(X, theta) - y) ** 2)
```

```python
# Calculate Initial Cost and Test Cost With Dummy Data Function
initial_cost = computeCost(X, Y, m, theta)
test_cost = computeCost(X, Y, m, np.c_[[-1, 2]])

print("Expected   Initial Cost:", 32.07)
print("Expected      Test Cost:", 54.24)

print("Calculated Initial Cost:", initial_cost)
print("Calculated    Test Cost:", test_cost)
```

```python
# Define Gradient Descent Algorithm
def gradientDescent(X, y, m, theta, alpha, num_iters):
    # Initial history to track cost changes 
    #    as gradient descent runs
    history = np.zeros(num_iters)
    
    # Perform gradient descent for num_iterations
    for i in range(num_iters):
        # Calculate the gradient
        gradient = np.matmul(X.T, (np.matmul(X, theta)) - y)
        
        # Update Theta (Parameters) Simultaneously
        theta -= (float(alpha) / m) * gradient
        
        # Save Current Cost in History
        history[i] = computeCost(X, y, m, theta)
        
    return (theta, history)
```

```python
%%time

# Find Parameters with Gradient Descent
theta = np.zeros((2, 1))
theta, history = gradientDescent(X, Y, m, theta, alpha, iterations)

# Print Found Parameters
for i in range(len(theta)):
    print("Theta_{0}: {1:10.7f}".format(i, theta[i][0]))

# Add Space before Execution Time Prints
print("\n")
```

```python
# Initialize Plot Object
fig = plt.figure(figsize=(12,8))

# Find Range and Extremes of History
min = np.min(history)
max = np.max(history)
delta = max - min

# Generate Cost History Plot
ax1 = fig.add_axes([0.08, 0, 0.42, 0.42])
ax1.set_title("Gradient Descent - Iteration vs. Cost")
ax1.set_xlabel("Iteration")
ax1.set_ylabel("Cost")
ax1.set_xlim([0, len(history)])
ax1.set_ylim([min - 0.1 * delta, max + 0.1 * delta])
ax1.plot(range(iterations), history)

# Find Range and Extremes of History
xtheta = range(4, 23)
ytheta = theta[0] + theta[1] * xtheta

# Generate Regression Line Plot
ax2 = fig.add_axes([0.58, 0, 0.42, 0.42])
ax2.set_title("Gradient Descent - Regression Output")
ax2.set_xlabel("Population of City in 10,000s")
ax2.set_ylabel("Profit in $10,000s")
ax2.scatter(X_Unbiased, Y, s=30, c="b", marker="x", linewidth=0.9)
ax2.plot(xtheta, ytheta, c="r", linewidth=0.8)
```

```python
# Regression Quality

# Predict Profit for a city with population of 35000
prediction = theta.T.dot([1, 3.5])[0]
print("A City of Population {0:5d} will have Approx ${1:9.2f} in Profit".format(35000, prediction * 10000))

# Predict Profit for a city with population of 70000
prediction = theta.T.dot([1, 7.0])[0]
print("A City of Population {0:5d} will have Approx ${1:9.2f} in Profit".format(70000, prediction * 10000))
```

```python
# Create grid coordinates for plotting
x = np.linspace(-10,10,100)
y = np.linspace(-1, 4, 100)
xx, yy = np.meshgrid(x, y)
Z = np.zeros((len(x), len(y)))

# Calculate Cost (Z-values) based on grid coefficients
for (i, j), value in np.ndenumerate(Z):
    Z[i,j] = computeCost(X, Y, m, [[xx[i,j]], [yy[i,j]]])
    
fig = plt.figure(figsize=(15,12))

# Generate Contour Plot
ax1 = fig.add_axes([0.08, 0, 0.42, 0.42])
ax1.set_title(r'Cost Contour - $\theta_0$ vs. $\theta_1$')
ax1.set_xlabel(r'$\theta_0$')
ax1.set_ylabel(r'$\theta_1$')
ax1.contour(xx, yy, Z, np.logspace(-2, 3, 20), cmap=plt.cm.jet)
ax1.scatter(theta[0], theta[1], c="purple")

# Generate Regression Line Plot
ax2 = fig.add_axes([0.58, 0, 0.42, 0.42], projection="3d")
ax2.set_title(r'Cost Surface - $\theta_0$ vs. $\theta_1$ vs. Cost')
ax2.set_xlabel(r'$\theta_0$')
ax2.set_ylabel(r'$\theta_1$')
ax2.set_zlabel("Cost")
ax2.set_zlim(Z.min(),Z.max())
ax2.view_init(elev=15, azim=230)
ax2.plot_surface(xx, yy, Z, rstride=1, cstride=1, alpha=0.6, cmap=plt.cm.jet)
```

```python
def featureNormalize(X):
    X_norm = np.copy(X)
    size = len(X[0])
    
    mean = np.zeros(size)
    std_dev = np.zeros(size)
    
    for i in range(size):
        mean[i] = np.mean(X_norm[:, i])
        std_dev[i] = np.std(X_norm[:, i], ddof=1)
        X_norm[:, i] = (X_norm[:, i] - mean[i]) / std_dev[i]
    
    return (X_norm, mean, std_dev)
```

```python
X, mean, std_dev = featureNormalize(X_NotNormalized)

print("First 10 Data Normalized Points")

for i in range(10):
    print("  x = [{0:9.5f} {1:9.5f}]".format(X[i, 0], X[i, 1]))
```
